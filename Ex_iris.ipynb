# Exercícios Iris
#### 1)Crie um loop para rodar o modelo knn com valores de K, entre 1 e 25.
#### 2)Faça o teste de performance para cada um dos valores de K e apresente os resultados

## Importando as bibliotecas

```python
from sklearn.datasets import load_iris #Importa o dataset iris
from sklearn.neighbors import KNeighborsClassifier #Importa o KNN
from sklearn.model_selection import train_test_split #Para dividir o dataset em treinamento e testes
from sklearn import metrics #Para visualizar a porcentagem de acerto
import matplotlib.pyplot as plt #Para a criação de um gráfico
```

## Carregar o dataset

```python
iris = load_iris()
```

## Inserir a "Observação" e o "Target"

```python
x = iris.data
y = iris.target
```

## Distribuição do Dataset

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25) #A divisão será de 25% dos arquivos para o treino
```

## Loop para rodar o modelo KNN

```python
numero = 1
valor_x = [] #Lista vazia para adicionar o numero a cada final (eixo X)
valor_y = [] #Lista vazia para adicionar a porcentagem de acertos (eixo y)

while numero <= 25:
    # Para fazer rodar o KNN com o número informado
    knn = KNeighborsClassifier(n_neighbors = numero) #Para selecionar o K
    knn.fit(x_train, y_train) #Para treinar a máquina
    
    previsoes = knn.predict(x_test) #Para fazer a predição
    acertos = metrics.accuracy_score(y_test, previsoes) #Para fazer a testagem na porcentagem

    #Para adicionar na lista
    valor_x.append(numero) 
    valor_y.append(acertos)
 
    numero += 1 #Adiciona sempre um número a mais até o 25
```

## Para demonstrar em gráfico

```python
plt.plot(valor_x, valor_y, marker='o')  # marker 'o' para mostrar os pontos
plt.title("Porcentagem de acerto KNN para diferentes valores de K")
plt.ylabel("K")
plt.xlabel("Porcentagem de Acerto")
plt.xticks(valor_x)  # mostra cada K no eixo x
plt.grid(True)
plt.show()
```
